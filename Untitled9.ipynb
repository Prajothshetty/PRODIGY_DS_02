```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('train.csv')

# See the first few rows
df.head()
```

```python
# Get a summary of the data (types and non-null counts)
df.info()

# See where we have missing values
print("\nMissing values before cleaning:")
print(df.isnull().sum())
```

```python
# Strategy 1: Fill missing 'Age' with the average (median) age
median_age = df['Age'].median()
df['Age'].fillna(median_age, inplace=True)

# Strategy 2: Drop the 'Cabin' column because it has too many missing values
df.drop('Cabin', axis=1, inplace=True)

# Strategy 3: Fill missing 'Embarked' with the most common port
most_common_port = df['Embarked'].mode()[0]
df['Embarked'].fillna(most_common_port, inplace=True)

# Check if we have any missing values left
print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Survived', data=df)
plt.title('Survival Count (0 = Died, 1 = Survived)')
plt.show()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival Count by Gender')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title('Survival Count by Passenger Class')
plt.legend(title='Passenger Class', labels=['1st', '2nd', '3rd'])
plt.show()
```

My analysis of the Titanic dataset revealed strong patterns. Survival rates were significantly higher for females compared to males. Furthermore, passengers in 1st class had a much higher chance of survival than those in 3rd class.
